```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import plotly.express as px


data=pd.read_csv('https://goo.gl/sp8gYx')

data.head(100)
```

**Masa Invariante En Unidades Naturales**

```python
E= (data['E1'] + data['E2'])**2
Px= (data['px1']+data['px2'])**2
Py= (data['py1']+data['py2'])**2
Pz= (data['pz1']+data['pz2'])**2
P = Px + Py + Pz
EP = E-P

m = np.sqrt(EP)

fig_1=px.histogram(m,m)
#fig_1.update_layout(title=r"$\text{Masa invariante} \ M, \ \text{%d bins}$" %1000, xaxis_title=r"$MeV$",yaxis_title="Eventos"
fig_1 .show()
```

```python
fig_1=px.histogram(data['M'], data['M'], nbins=1000)
fig_1.show()
```
