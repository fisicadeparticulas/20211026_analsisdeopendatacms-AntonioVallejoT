```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import plotly.express as px


data=pd.read_csv('https://goo.gl/sp8gYx')

data.head(100)
```

```python
fig_1=px.histogram(data['pt1'],data['pt1'],nbins=10000, range_x=["0","100"],labels={"x": "$P_{T1}$" }  )
fig_1.update_layout(title=r"$\text{Momento transverso} P_{T_1}$")
fig_1.show()

fig_11=px.histogram(data['pt2'],data['pt2'],nbins=10000, range_x=["0","100"],labels={"x": "$P_{T2}$" }  )
fig_1.update_layout(title=r"$\text{Momento transverso} P_{T_2}$")
fig_11.show()
```

```python
fig_1=px.histogram(data['phi1'], data['phi1'], nbins=100,  labels={"x": "radianes", } )
fig_1.update_layout(title=r"$\phi_1$")
fig_1.show()

fig_11=px.histogram(data['phi2'], data['phi2'], nbins=100,  labels={"x": "radianes", } )
fig_11.update_layout(title=r"$\phi_1$")
fig_11.show()

#dphi = np.abs(data['phi2'] - data['phi1'])
#fig=px.line(dphi,y=dphi)
#fig_12.update_layout(title=r"d $\phi$")
#fig_12.show()

```
